```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Bidirectional, Dropout, Dense, Layer
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

print("TensorFlow version:", tf.__version__)
```

```python
# 1. Create Synthetic Stock Data with a realistic trend
np.random.seed(42)
time_steps = 60
num_days = 1200

# Generating a random walk to mimic real stock trends instead of pure white noise
steps = np.random.normal(0, 2, num_days)
close_prices = 100 + np.cumsum(steps)
open_prices = close_prices + np.random.normal(0, 1, num_days)
high_prices = np.maximum(open_prices, close_prices) + np.random.uniform(0, 3, num_days)
low_prices = np.minimum(open_prices, close_prices) - np.random.uniform(0, 3, num_days)
volume = np.random.uniform(10000, 500000, num_days)

df = pd.DataFrame({
    'Open': open_prices, 'High': high_prices, 'Low': low_prices, 
    'Close': close_prices, 'Volume': volume
})

# 2. Scale Features (X) and Target (Y) SEPARATELY to avoid inverse transform distortion
scaler_X = MinMaxScaler(feature_range=(0, 1))
scaler_Y = MinMaxScaler(feature_range=(0, 1))

# Fit on all features
scaled_features = scaler_X.fit_transform(df)
# Fit explicitly on the Close price
scaled_target = scaler_Y.fit_transform(df[['Close']])

# 3. Sliding Window Creation
def build_supervised_sequences(features, target, seq_length=60):
    X, y = [], []
    for i in range(len(features) - seq_length):
        X.append(features[i : i + seq_length])
        y.append(target[i + seq_length])
    return np.array(X), np.array(y)

X, y = build_supervised_sequences(scaled_features, scaled_target, seq_length=time_steps)

# 4. Train/Test Split (80/20)
split = int(len(X) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

print(f"X_train Shape: {X_train.shape} | y_train Shape: {y_train.shape}")
```

```python
class AttentionLayer(Layer):
    def __init__(self, **kwargs):
        super(AttentionLayer, self).__init__(**kwargs)

    def build(self, input_shape):
        hidden_dim = input_shape[-1]
        self.W_a = self.add_weight(name='att_weight', shape=(hidden_dim, hidden_dim),
                                   initializer='glorot_uniform', trainable=True)
        self.b_a = self.add_weight(name='att_bias', shape=(hidden_dim,),
                                   initializer='zeros', trainable=True)
        super(AttentionLayer, self).build(input_shape)

    def call(self, h_t):
        # Score calculation: e_t = tanh(W_a * h_t + b_a)
        score = tf.tanh(tf.tensordot(h_t, self.W_a, axes=[[2], [0]]) + self.b_a)
        score = tf.reduce_mean(score, axis=-1) 
        
        # Softmax normalization
        alpha_t = tf.nn.softmax(score, axis=1)
        alpha_t_expanded = tf.expand_dims(alpha_t, axis=-1)
        
        # Context Vector context = sum(alpha_t * h_t)
        context_vector = tf.reduce_sum(alpha_t_expanded * h_t, axis=1)
        return context_vector, alpha_t

def build_optimized_network(time_steps, num_features):
    inputs = Input(shape=(time_steps, num_features))
    
    # BiLSTM Layer 1 (Extracts sequential context)
    bilstm_1 = Bidirectional(LSTM(64, return_sequences=True))(inputs)
    dropout_1 = Dropout(0.2)(bilstm_1)
    
    # Attention Block
    context_vector, _ = AttentionLayer()(dropout_1)
    
    # Fully Connected Regression Block for variance retention
    dense_1 = Dense(32, activation='relu')(context_vector)
    outputs = Dense(1, activation='linear')(dense_1)
    
    model = Model(inputs=inputs, outputs=outputs)
    # Using Huber loss or MSE with a finely tuned learning rate
    model.compile(optimizer=Adam(learning_rate=0.001), loss='huber')
    return model

model = build_optimized_network(time_steps=60, num_features=5)
```

```python
# Using Early Stopping to prevent the model from getting stuck in a mean-prediction state
early_stop = EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=40, # Increased epochs for complex pattern convergence
    batch_size=32,
    callbacks=[early_stop],
    verbose=1
)
```

```python
# 1. Predict
predictions_scaled = model.predict(X_test)

# 2. Seamless Inverse Transform using the dedicated target scaler
predictions_actual = scaler_Y.inverse_transform(predictions_scaled)
y_test_actual = scaler_Y.inverse_transform(y_test)

# 3. Metrics Comparison
rmse = np.sqrt(mean_squared_error(y_test_actual, predictions_actual))
mae = mean_absolute_error(y_test_actual, predictions_actual)
r2 = r2_score(y_test_actual, predictions_actual)

print(f"\n--- PERFORMANCE METRICS ---")
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
print(f"R2 Score (Should be close to 1): {r2:.4f}")

# 4. Plotting
plt.figure(figsize=(15, 6))
plt.plot(y_test_actual, label='Actual Close Price', color='#008080', linewidth=2)
plt.plot(predictions_actual, label='Optimized Prediction', color='#FF4500', linestyle='--', linewidth=2)
plt.title('Corrected Stock Price Prediction (Scale and Variance Aligned)', fontsize=14)
plt.xlabel('Timeline indices (Test Split)', fontsize=12)
plt.ylabel('Stock Value ($)', fontsize=12)
plt.legend(loc='upper left')
plt.grid(True, alpha=0.3)
plt.show()
```
